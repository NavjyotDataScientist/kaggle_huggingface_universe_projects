RL is learning by trial and error to maximize rewards.

SIMPLEST POSSIBLE RL EXAMPLE (NO GYM, NO MAGIC)
üéØ Goal:

Reach number 10 starting from 0

```python
import numpy as np

# Environment
goal = 10
states = range(0, 11)
actions = [-1, 1]

# Q-table
Q = np.zeros((11, 2))  # states x actions

# Hyperparameters
alpha = 0.1
gamma = 0.9
episodes = 200

for episode in range(episodes):
    state = 0  # start
    done = False

    while not done:
        # choose action
        action_idx = np.random.choice([0, 1])
        action = actions[action_idx]

        next_state = max(0, min(10, state + action))

        # reward
        reward = 10 if next_state == goal else -1

        # Q update
        Q[state, action_idx] += alpha * (
            reward + gamma * np.max(Q[next_state]) - Q[state, action_idx]
        )

        state = next_state
        if state == goal:
            done = True

# Test
state = 0
path = [state]
while state != goal:
    action = actions[np.argmax(Q[state])]
    state = max(0, min(10, state + action))
    path.append(state)

print("Learned path:", path)
```
