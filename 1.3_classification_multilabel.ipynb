Real-world examples:

| Movie Description     | Action | Comedy | Drama |
| --------------------- | ------ | ------ | ----- |
| Hero saves the city   | 1      | 0      | 0     |
| Funny love story      | 0      | 1      | 1     |
| War with emotions     | 1      | 0      | 1     |
| Romantic comedy movie | 0      | 1      | 0     |


```python
# 1. Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import classification_report

# 2. Create dataset
data = {
    "text": [
        "hero saves city",
        "funny love story",
        "war and emotions",
        "romantic comedy movie"
    ],
    "Action": [1, 0, 1, 0],
    "Comedy": [0, 1, 0, 1],
    "Drama": [0, 1, 1, 0]
}

df = pd.DataFrame(data)

# 3. Features and labels
X = df["text"]
y = df[["Action", "Comedy", "Drama"]]

# 4. Text to numerical
vectorizer = CountVectorizer()
X_vec = vectorizer.fit_transform(X)

# 5. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_vec, y, test_size=0.3, random_state=42
)

# 6. Train multilabel model
model = OneVsRestClassifier(LogisticRegression())
model.fit(X_train, y_train)

# 7. Predictions
y_pred = model.predict(X_test)

# 8. Evaluation
print(classification_report(y_test, y_pred))
```
