Definition:

Data collected over time where order matters

Example:

Stock price every day

Temperature every hour

Sales every month

If you shuffle rows → model breaks ❌

```python
# Time Series - Lag Feature Example (One Cell)

import pandas as pd
from sklearn.linear_model import LinearRegression

# Create simple time series data
data = {
    "Day": [1, 2, 3, 4, 5, 6],
    "Sales": [10, 12, 13, 15, 18, 20]
}

df = pd.DataFrame(data)

# Create lag feature
df["Sales_lag_1"] = df["Sales"].shift(1)

# Drop first row (NaN lag)
df = df.dropna()

X = df[["Sales_lag_1"]]
y = df["Sales"]

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict next day
next_day_sales = model.predict([[20]])

print("Predicted next day sales:", next_day_sales[0])
```

```python
# Lag feature = past value used as an input

# In simple words:

# Yesterday helps predict today
```
