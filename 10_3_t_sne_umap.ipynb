 t-SNE REALLY DOES (ONE LINE)

 t-SNE says:

“I’ll draw a map where similar songs sit near each other.”

```python
# 1️⃣ Imports
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# 2️⃣ Load data
digits = load_digits()
X = digits.data
y = digits.target

# 3️⃣ Scale (important for t-SNE)
X_scaled = StandardScaler().fit_transform(X)

# 4️⃣ Apply t-SNE (to 2D)
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# 5️⃣ Plot
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y)
plt.colorbar(label="Digit Label")
plt.title("t-SNE Visualization of Digits Dataset")
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.show()
```

WHAT UMAP DOES (ONE LINE)

UMAP places similar points close together like t-SNE but faster and preserves more global structure.

Great for visualization

Can handle larger datasets than t-SNE

Also non-linear

```python
# 1️⃣ Imports
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
import umap

# 2️⃣ Load data
digits = load_digits()
X = digits.data
y = digits.target

# 3️⃣ Scale
X_scaled = StandardScaler().fit_transform(X)

# 4️⃣ Apply UMAP
reducer = umap.UMAP(n_components=2, random_state=42)
X_umap = reducer.fit_transform(X_scaled)

# 5️⃣ Plot
plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y)
plt.colorbar(label="Digit Label")
plt.title("UMAP Visualization of Digits Dataset")
plt.xlabel("UMAP 1")
plt.ylabel("UMAP 2")
plt.show()
```
