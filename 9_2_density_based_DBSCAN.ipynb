DBSCAN groups points that are packed closely together and marks isolated points as noise.

Groups of people sitting close → clusters

A person sitting alone far away → noise

No need to ask: “How many groups?” ❌



```python
# 1️⃣ Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# 2️⃣ Dataset
X = np.array([
    [1, 1],
    [1, 2],
    [2, 1],
    [8, 8],
    [8, 9],
    [9, 8],
    [15, 15]  # noise
])

# 3️⃣ DBSCAN Model
dbscan = DBSCAN(eps=1.5, min_samples=2)
labels = dbscan.fit_predict(X)

# 4️⃣ Plot
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("DBSCAN Clustering (-1 = Noise)")
plt.show()
```
