```python
# GRADIENT BOOSTING â€” INTUITION (30 SECONDS)

# Think like this:

# First model predicts something

# It makes mistakes

# Second model learns only the mistakes

# Third model learns remaining mistakes

# Final prediction = sum of all corrections

# ðŸ‘‰ Like fixing an exam paper question by question, not rewriting the whole paper.
```

```python
# Gradient Boosting Intuition - One Cell Example

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Create simple regression data
X, y = make_regression(
    n_samples=200,
    n_features=1,
    noise=15,
    random_state=42
)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Gradient Boosting Model
gbr = GradientBoostingRegressor(
    n_estimators=100,    # number of weak learners
    learning_rate=0.1,   # how much each tree contributes
    max_depth=2          # weak tree depth
)

# Train
gbr.fit(X_train, y_train)

# Predict
pred = gbr.predict(X_test)

# Evaluation
print("MSE:", mean_squared_error(y_test, pred))
```

```python
# WHAT TREES LEARN IN CLASSIFICATION?

# They donâ€™t predict numbers directly.
# They learn to separate classes better by focusing on:

# Hard samples

# Previously misclassified points
```

```python
# Gradient Boosting Classifier - One Cell Example

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Create classification data
X, y = make_classification(
    n_samples=300,
    n_features=4,
    n_informative=2,
    random_state=42
)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3
)

# Gradient Boosting Classifier
gbc = GradientBoostingClassifier(
    n_estimators=100,    # number of weak trees
    learning_rate=0.1,   # step size
    max_depth=2          # weak learner depth
)

# Train
gbc.fit(X_train, y_train)

# Predict
pred = gbc.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, pred))
```
