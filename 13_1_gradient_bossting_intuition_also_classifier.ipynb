WHAT IS XGBOOST (IN ONE LINE)

XGBoost = Gradient Boosting, but engineered for speed, stability, and scale.

Same idea you already know â€” better execution.

ðŸ” WHY SKLEARN GRADIENT BOOSTING WAS NOT ENOUGH

Sklearn GB:

Slow

No parallelization

Overfits easily

No missing value handling

XGBoost fixes all of this.

```python
# Construction team

# Gradient Boosting = good workers

# XGBoost = good workers + machines + supervisor + safety rules

# Same work, better system.
```

```python
# XGBoost Classifier - One Cell Example

from xgboost import XGBClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Create dataset
X, y = make_classification(
    n_samples=300,
    n_features=4,
    n_informative=2,
    random_state=42
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3
)

# XGBoost Model
model = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    eval_metric='logloss'
)

# Train
model.fit(X_train, y_train)

# Predict
pred = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, pred))
```

LightGBM

ðŸŒ¿ Leaf-wise tree growth

Thatâ€™s the heart of the difference.

```python
# XGBoost tree:
#         root
#       /      \
#    level    level
#    /   \    /   \

# LightGBM tree:
#         root
#            \
#              \
#                \
#             (deep on hardest path)
```

```python
# LightGBM Classifier - One Cell Example

from lightgbm import LGBMClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Create dataset
X, y = make_classification(
    n_samples=300,
    n_features=10,
    n_informative=4,
    random_state=42
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3
)

# LightGBM Model
model = LGBMClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=-1  # leaf-wise growth
)

# Train
model.fit(X_train, y_train)

# Predict
pred = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, pred))
```

```python
# CATBOOST SOLVES THIS AUTOMATICALLY
# ðŸ’¡ Core idea:

# CatBoost handles categorical features internally

# You give raw text â†’ CatBoost learns safely.
```

```python
# CatBoost Classifier - One Cell Example

from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd

# Sample dataset with categorical features
data = {
    "City": ["Delhi", "Mumbai", "Paris", "Delhi", "Paris"],
    "Gender": ["M", "F", "F", "M", "M"],
    "Age": [25, 30, 22, 35, 28],
    "Purchased": [1, 0, 1, 0, 1]
}

df = pd.DataFrame(data)

X = df.drop("Purchased", axis=1)
y = df["Purchased"]

# Identify categorical columns
cat_features = ["City", "Gender"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3
)

# Model
model = CatBoostClassifier(
    iterations=100,
    learning_rate=0.1,
    verbose=False
)

# Train (no encoding needed!)
model.fit(X_train, y_train, cat_features=cat_features)

# Predict
pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, pred))
```
