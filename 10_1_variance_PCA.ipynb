PCA ‚Äî EXPLAINED VARIANCE

(How much information PCA keeps)

This answers the question:

‚ÄúHow many components are enough?‚Äù

A movie

You summarize it in:

1 sentence ‚Üí bad

5 sentences ‚Üí decent

10 sentences ‚Üí very good

üß† Explained variance =
‚ÄúHow much of the original movie did we keep?‚Äù

```python
# 1Ô∏è‚É£ Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 2Ô∏è‚É£ Sample dataset (3 features)
X = np.array([
    [160, 60, 30],
    [165, 65, 32],
    [170, 70, 35],
    [175, 75, 37],
    [180, 80, 40]
])

# 3Ô∏è‚É£ Standardize
X_scaled = StandardScaler().fit_transform(X)

# 4Ô∏è‚É£ PCA with all components
pca = PCA()
pca.fit(X_scaled)

# 5Ô∏è‚É£ Explained variance
explained = pca.explained_variance_ratio_

# 6Ô∏è‚É£ Plot
plt.bar(range(1, len(explained)+1), explained)
plt.xlabel("Principal Components")
plt.ylabel("Explained Variance Ratio")
plt.title("PCA Explained Variance")
plt.show()

# 7Ô∏è‚É£ Print values
for i, var in enumerate(explained, 1):
    print(f"PC{i}: {var:.2%}")
```
