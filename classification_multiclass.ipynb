Multiclass Classification means:

Predicting ONE class out of MORE than two classes.

Examples:

Digit recognition ‚Üí 0‚Äì9

Email category ‚Üí Primary / Social / Promotions

News ‚Üí Sports / Politics / Tech / Business

üëâ Output is one label, but more than 2 options.

key Difference (Binary vs Multiclass)
Type	Classes
Binary Classification	2 classes
Multiclass Classification	3 or more classes

üìå Logic stays the same.
üìå Only number of classes changes.

Real-World Project Example
üìå Project Title

Handwritten Digit Recognition (0‚Äì9)

Goal:
Given an image of a digit, predict which digit it is (0‚Äì9).

```python
# 1. Import libraries
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# 2. Load dataset
digits = load_digits()
X = digits.data      # features (pixel values)
y = digits.target    # labels (0‚Äì9)

# 3. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 4. Train model (Multiclass classifier)
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)

# 5. Predictions
y_pred = model.predict(X_test)

# 6. Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
