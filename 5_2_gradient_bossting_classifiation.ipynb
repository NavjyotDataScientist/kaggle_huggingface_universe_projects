1️⃣ WHY Gradient Boosting Classification EXISTS

You already know:

Decision Tree → learns rules (but overfits)

Random Forest → many trees vote (stable)

Now Gradient Boosting asks a different question:

“What if each new tree FIXES the mistakes of the previous one?”

That idea creates Gradient Boosting.



```python
# “Gradient Boosting Classification builds models sequentially, where each new tree corrects the errors made by previous ones.”
```

```python
# ==================================
# Gradient Boosting Classification
# ==================================

# 1. Import libraries
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report

# 2. Create dataset
data = {
    "Hours_Studied": [1, 2, 3, 4, 5, 6, 7, 8],
    "Pass": [0, 0, 0, 1, 1, 1, 1, 1]
}

df = pd.DataFrame(data)
print("Dataset:")
print(df)

# 3. Feature & target
X = df[["Hours_Studied"]]
y = df["Pass"]

# 4. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 5. Train Gradient Boosting model
model = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)
model.fit(X_train, y_train)

# 6. Predict
y_pred = model.predict(X_test)

# 7. Evaluate
accuracy = accuracy_score(y_test, y_pred)

print("\nModel Accuracy:", accuracy)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
