```python
# K-Means groups similar data points into K clusters.
```

```python
# 1Ô∏è‚É£ FIRST: WHAT CHANGED?

# So far:

# You had labels (y)

# Model learned to predict y

# Now:

# No labels

# Model finds structure by itself

# üëâ This is Unsupervised Learning.
```

```python
# 1Ô∏è‚É£ Imports
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 2Ô∏è‚É£ Dataset
X, _ = make_blobs(
    n_samples=300,
    centers=3,
    cluster_std=0.6,
    random_state=42
)

# 3Ô∏è‚É£ Model
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# 4Ô∏è‚É£ Visualization
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    marker="X"
)
plt.title("K-Means Clustering")
plt.show()
```

```python
# 1Ô∏è‚É£ Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 2Ô∏è‚É£ Very simple dataset
X = np.array([
    [1, 1],
    [1, 2],
    [2, 1],
    [8, 8],
    [8, 9],
    [9, 8]
])

# 3Ô∏è‚É£ K-Means with K=2
kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(X)

# 4Ô∏è‚É£ Plot
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    marker='X'
)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Simple K-Means Example")
plt.show()
```
