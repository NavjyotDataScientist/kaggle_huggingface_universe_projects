WHAT PCA REALLY DOES (ONE LINE)

PCA compresses data by keeping the most important directions of variation.

No prediction.
No labels.
Only information preservation.

```python
# 1️⃣ Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 2️⃣ Dataset (Height, Weight)
X = np.array([
    [160, 60],
    [165, 65],
    [170, 70],
    [175, 75],
    [180, 80]
])

# 3️⃣ Standardize
X_scaled = StandardScaler().fit_transform(X)

# 4️⃣ Apply PCA (reduce 2D → 1D)
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X_scaled)

# 5️⃣ Plot
plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
plt.xlabel("Height (scaled)")
plt.ylabel("Weight (scaled)")
plt.title("Before PCA")
plt.show()

plt.scatter(X_pca, np.zeros_like(X_pca))
plt.xlabel("Principal Component 1")
plt.title("After PCA (2D → 1D)")
plt.show()
```
