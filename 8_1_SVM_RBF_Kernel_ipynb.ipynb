A kernel changes the shape of the decision boundary without explicitly transforming data.

You don’t do the math — SVM does it implicitly.

```python
# | Kernel  | When to Use                |
# | ------- | -------------------------- |
# | linear  | Simple, fast               |
# | poly    | Polynomial patterns        |
# | **rbf** | Default, curved boundaries |
# | sigmoid | Rare                       |
```

```python
# 1️⃣ Imports
from sklearn.datasets import make_circles
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 2️⃣ Non-linear dataset
X, y = make_circles(
    n_samples=200,
    noise=0.1,
    factor=0.4,
    random_state=42
)

# 3️⃣ Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 4️⃣ Model (RBF Kernel)
model = SVC(kernel="rbf", C=1, gamma="scale")
model.fit(X_train, y_train)

# 5️⃣ Prediction
y_pred = model.predict(X_test)

# 6️⃣ Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
```
